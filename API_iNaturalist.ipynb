```python
import os
import requests

ruta = '/content/Opiliones'

!mkdir {ruta}
```

```python
contador = 4785
print('Contador con valor', contador)
```

```python
API_url = 'https://api.inaturalist.org/v1/observations'

params = {
    "taxon_id": 47367,
    "quality_grade": "needs_id",
    "per_page": 200,
    "page": 13
 }
```

```python
response = requests.get( API_url, params=params )

if response.status_code == 200:
  data = response.json()
  for observacion in data['results']:
    imagenes = observacion['photos']
    for imagen in imagenes:
      imagen_url = imagen['url'].replace("square", "large")
      imagen_respuesta = requests.get(imagen_url)
      file = os.path.join(f'{ruta}', f'{contador}.jpg')
      contador += 1
      with open(file, 'wb') as archivo:
        archivo.write(imagen_respuesta.content)
else:
  print('Error al buscar observaciones')
```

```python
num_imagenes = 0

for archivo in os.listdir(ruta):
    num_imagenes += 1

print(f'Número total de imágenes en la carpeta: {num_imagenes} contador: {contador}')
```
